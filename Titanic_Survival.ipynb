## **Import Required Libraries**

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# from google.colab import drive
# drive.mount('/content/drive/')
```

**Reading Titanic_train.csv**

```python
titanic = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Titanic_train.csv')
titanic.head()
```

**Separate the categorical Data**

```python
titanic_cat = titanic.select_dtypes(object)
titanic_cat.head()
```

```python
titanic_cat.isnull().sum()
```

**Categorical Data Preparation**

```python
titanic_cat.drop(['Name', 'Ticket'], axis = 1, inplace = True)
titanic_cat.head(10)
```

```python
titanic_cat.isnull().sum()
```

**Replace NaN values of Cabin with max count**

**Replace NaN values of Embarked with max count**

```python
titanic_cat.Cabin.fillna(titanic_cat.Cabin.value_counts().idxmax(), inplace = True)
titanic_cat.head(10)
```

```python
titanic_cat.Embarked.fillna(titanic_cat.Embarked.value_counts().idxmax(), inplace = True)
titanic_cat.head(10)
```

```python
titanic_cat.isnull().sum()
```

```python
titanic_cat.describe()
```

**Separate Numerical Data**

```python
titanic_num = titanic.select_dtypes(np.number)
titanic_num.head(10)
```

```python
titanic_num.isnull().sum()
```

**Numerical Data Preparation**

```python
titanic_num.drop(['PassengerId'], axis = 1, inplace = True)
titanic_num.head(10)
```

```python
titanic_num.isnull().sum()
```

**Replace NaN values of Age with max count**

```python
titanic_num.Age.fillna(titanic_num.Age.mean(), inplace = True)
titanic_num.head(10)
```

# **Labeling the Categorical Data**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
lab_en = LabelEncoder()
titanic_cat = titanic_cat.apply(lab_en.fit_transform)
titanic_cat.head(10)
```

**Merge Preprocessed Categorical and Numericals datas**

```python
titanic_final = pd.concat([titanic_cat, titanic_num], axis = 1)
titanic_final.head(10)
```

**Prepare Input-Output Data**

```python
X = titanic_final.drop(['Survived'], axis = 1)
X.head(10)
```

```python
Y = titanic_final['Survived']
Y.head(10)
```

**Fetch Training Inputs and Outputs**

```python
X_train = np.array(X[0:(int)(len(X)*0.98)])
X.head(10)
```

```python
Y_train = np.array(Y[0:(int)(len(Y)*0.98)])
Y.head(10)
```

**Fetch Testing Inputs and Outputs**

```python
X_test = np.array(X[(int)(len(X)*0.98):])
X.head(10)
```

```python
Y_test = np.array(Y[(int)(len(Y)*0.98):])
Y.head(10)
```

**Model Selection/Object Creation**

```python
LR = LogisticRegression()
KNN = KNeighborsClassifier()
```

**Fitting the Model (with Training Data)**

```python
LR_fit = LR.fit(X_train, Y_train)
KNN_fit = KNN.fit(X_train, Y_train)
```

**Predicting the output (with Testing Data)**

```python
LR_predict = LR_fit.predict(X_test)
KNN_predict = KNN_fit.predict(X_test)
```

**Accuracy Calculation**

```python
LR_accuracy = accuracy_score(LR_predict, Y_test) * 100
LR_accuracy
```

```python
KNN_accuracy = accuracy_score(KNN_predict, Y_test) * 100
KNN_accuracy
```
